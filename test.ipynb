```python
import numpy as np
import matplotlib.pyplot as plt

def get_rope_clocks(position, head_dim=64, theta_base=10000):
    # 1. Calculate the 'Speeds' (Frequencies) for 32 pairs
    indices = np.arange(0, head_dim, 2)
    inv_freq = 1.0 / (theta_base ** (indices / head_dim))
    
    # 2. Calculate the 'Angles' for this specific word position
    angles = position * inv_freq
    return angles

# Compare Word #1 and Word #50
pos1_angles = get_rope_clocks(position=1)
pos50_angles = get_rope_clocks(position=50)

# Let's look at the first pair (Fast) vs last pair (Slow)
print(f"Clock 1 (Fastest) - Word 1: {np.degrees(pos1_angles[0]):.1f}°")
print(f"Clock 1 (Fastest) - Word 50: {np.degrees(pos50_angles[0]):.1f}° (Spun many times!)")
print("-" * 30)
print(f"Clock 32 (Slowest) - Word 1: {np.degrees(pos1_angles[-1]):.4f}°")
print(f"Clock 32 (Slowest) - Word 50: {np.degrees(pos50_angles[-1]):.4f}° (Barely moved)")
```
